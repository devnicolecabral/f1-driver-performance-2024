Neste notebook, vamos analisar o desempenho das equipes (constructors) na temporada de 2024. Vamos usar os dados de resultados das corridas para somar os pontos conquistados por cada equipe, criando uma vis√£o geral da classifica√ß√£o atual.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Estilo dos gr√°ficos

```python
sns.set(style='darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 2. Lendo os dados filtrados e de apoio

```python
constructor_results_2024 = pd.read_csv('../data_filtrada/constructor_results_2024.csv')
constructors = pd.read_csv('../data/constructors.csv')
```

## 3. Preparando os dados

```python
# üîó Juntando resultados com os dados das equipes
resultados_constructors = constructor_results_2024.merge(constructors, on='constructorId', how='left')

# üßÆ Somando os pontos por equipe
pontuacao_constructors = resultados_constructors.groupby('name')['points'].sum().sort_values(ascending=False).reset_index()
```

## 4. Visualiza√ß√£o dos dados

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=pontuacao_constructors, x='points', y='name', palette='Blues_r')
plt.title('Pontua√ß√£o das Equipes - Temporada 2024', fontsize=16)
plt.xlabel('Pontos')
plt.ylabel('Equipe')
plt.tight_layout()
plt.show()
```

## 5. Conclus√£o

Neste notebook, realizamos a an√°lise da pontua√ß√£o das equipes na temporada de 2024. A partir dos dados de `constructor_results_2024.csv` e `constructors.csv`, conseguimos calcular os pontos totais por equipe e visualizar os resultados com um gr√°fico de barras.

- üß† Essa an√°lise nos ajuda a ver o desempenho das principais equipes.
